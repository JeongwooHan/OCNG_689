```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# Load a stream flow data distributed by USGS

f = open('merrimack_discharge.dat')
f.seek(0)

data = []
date = []
for line in f.readlines():
    line_data = line.split('\t')
    if not line_data: continue
    if line_data[0] == 'USGS':
        date_temp = line_data[2].split('-')
        date_line = [int(date_temp[0]), int(date_temp[1]), int(date_temp[2])]
        data_line = float(line_data[3]) * 0.028316 # To convert unit from cfs to cms
        date.append(date_line)
        data.append(data_line)
f.close()

data_vec = np.hstack((np.array(date), np.array(data).reshape(len(data),1)))
```

```python
from datetime import datetime, timedelta

yr_in_data = np.unique(data_vec[:,0]) # Figure out which years are in the file

def date_frame(year):
    do = datetime(year, 1, 1)
    dt = timedelta(days = 1)

    dates = do

    years = []
    months = []
    dayss= []

    n = 0

    while dates.year == year:
        dates = do + n*dt
        years.append(dates.year)
        months.append(dates.month)
        dayss.append(dates.day)
        n +=1
    years = np.array(years)
    months = np.array(months)
    dayss = np.array(dayss)

    date_vec = np.hstack((np.hstack((years[:,np.newaxis], months[:,np.newaxis])), dayss[:, np.newaxis]))
    date_vec = np.delete(date_vec,-1, axis = 0)
    
    if date_vec[:,1].shape[0] == 366:
        pass
    else:
        line_num = 0
    
        for line in date_vec:
            if (line[1] == 2) and (line[2] == 28):
                break
            line_num += 1

        before_29 = date_vec[0:line_num+1,:]

        add_line = np.array([int(year), 2, 29])

        add_line= add_line.reshape(1,3)

        after_29 = date_vec[line_num+1:,:]

        date_vec = np.append(before_29, add_line, axis = 0)

        date_vec = np.append(date_vec, after_29, axis = 0)

    
    return date_vec
```

```python
# Calculate Flow mean & Standard deviation
temp_date = date_frame(year = 1936)

all_data_vec = np.hstack((temp_date, np.ones((366, len(yr_in_data)))*float('nan')))

for yr in yr_in_data:
    year_data = data_vec[np.where(data_vec[:,0] == yr)[0],:]
    month_in_year = np.unique(year_data[:,1])
    
    for month in month_in_year:
        if month == 2:
            if len(all_data_vec[np.where(temp_date[:,1] == month)[0], int(yr-1923+1+2)]) != len(year_data[np.where(year_data[:,1] == month)[0], -1]):
                all_data_vec[np.where(temp_date[:,1] == month)[0][0:-1], int(yr-1923+1+2)] = year_data[np.where(year_data[:,1] == month)[0], -1]
            else:
                all_data_vec[np.where(temp_date[:,1] == month)[0], int(yr-1923+1+2)] = year_data[np.where(year_data[:,1] == month)[0], -1]
        else:
            all_data_vec[np.where(temp_date[:,1] == month)[0], int(yr-1923+1+2)] = year_data[np.where(year_data[:,1] == month)[0], -1]      
            
flow_mean = np.nanmean(all_data_vec[:,3:], axis= 1)
flow_std = np.nanstd(all_data_vec[:,3:], axis= 1)

upper_bound = flow_mean + flow_std
lower_bound = flow_mean - flow_std
```

```python
# Extract flow data in a specific year

years = np.array([1936, 1938, 1960, 1987])

flow_in_year = {} # This dictionary will contain keys of year and values of flowdata[yr,mon,day,val]

for year in years:
    yr_idx = np.where(data_vec[:,0] == year)
    flow_in_year['%s' %year]= data_vec[yr_idx,:]


# Calculate mean flow and standard deviation
# --- 
# 1936 and 1960 are leap year. So the difference in data size put some troubles in calculation.
# Intetionallay make data of 1938 and 1987 include 'NaN' on 29the days of Feb in two data arrays

# 1) Function to Add 29th day
def add_29(yr):
    line_num = 0
    for line in flow_in_year[yr][0]:
        if (line[1] == 2) and (line[2] == 28):
            break
        line_num += 1
    
    before_29 = flow_in_year[yr][0][0:line_num+1,:]

    add_line = np.array([int(yr), 2, 29, float('NaN')])
    
    add_line= add_line.reshape(1,4)
    
    after_29 = flow_in_year[yr][0][line_num+1:,:]
    
    new_name = yr + 'new'
    
    flow_in_year[new_name]= np.append(before_29, add_line, axis = 0)
 
    flow_in_year[new_name] = np.append(flow_in_year[new_name], after_29, axis = 0)
    
    return flow_in_year[new_name]


#2) flow data in each year    

yr1 = '1938'

flow_1938 = add_29(yr = yr1)

yr2 = '1987'
flow_1987 = add_29(yr = yr2)

flow_1936 = flow_in_year['1936'][0]

flow_1960 = flow_in_year['1960'][0]

#3) Mean flow and standard deviation

num_yr = 4

flow_vec = np.empty((flow_1936.shape[0], num_yr), dtype = float)

flow_vec[:,0] = flow_1936[:,-1]
flow_vec[:,1] = flow_1938[:,-1]
flow_vec[:,2] = flow_1960[:,-1]
flow_vec[:,3] = flow_1987[:,-1]
  
```

```python
from datetime import datetime
import matplotlib.dates
from matplotlib import colors

dates = []
for itr in range(0, len(flow_in_year['1936'][0][:,0])):
    date = datetime(int(flow_in_year['1936'][0][itr,0]),int(flow_in_year['1936'][0][itr,1]),int(flow_in_year['1936'][0][itr,2]))
    dates.append(date)
    
# Plot data
fig = plt.figure(figsize=(15,5))
ax = fig.add_subplot(1,1,1)
ax.plot(dates, flow_mean,'k-', lw =2.5, label = 'Mean')
ax.plot(dates, flow_1936[:,-1],'r-', lw =3, label = '1936')
ax.plot(dates, flow_1938[:,-1],'m--', lw =3, label = '1938')
ax.plot(dates, flow_1960[:,-1],'-.', color = 'orange', lw =3, label = '1960')
ax.plot(dates, flow_1987[:,-1],'g:', color = 'darkgreen', lw =3, label = '1987')
ax.fill_between(dates, lower_bound, upper_bound, alpha = 0.25, edgecolor = 'gray', facecolor = 'gray')
ax.grid(True, color = 'k', linestyle = ':', lw=0.25)
ax.xaxis.set_major_formatter(matplotlib.dates.DateFormatter('%b'))
ax.tick_params(axis = 'x', labelsize = 15)
ax.tick_params(axis = 'y', labelsize = 15)
ax.set_xticks(dates, minor = True)
ax.set_ylabel('Discharge [ m$^{3}$ s$^{-1}$ ]', fontsize = 15)
ax.set_title('Merrimack River Floods at Lowell, MA', fontsize = 15)
ax.spines['top'].set_linewidth(2)
ax.spines['bottom'].set_linewidth(2)
ax.spines['right'].set_linewidth(2)
ax.spines['left'].set_linewidth(2)
ax.set_ylim([-100,np.nanmax(flow_vec)])
ax.legend(loc = 2)
```

```python
def fileopen(file):
    ''' Open file and extract data values'''
    f = open(file)
    f.seek(0)

    i = 0
    for line in f.readlines():
        data = line.split('\t')
        if data[0] == '*END*\n':
            str_end = i
        i +=1
    
    f.close()
    f = open(file)

    data = f.readlines()[str_end+1:]
    
    return data

def data_tsd(data):
    '''Collect temperature, salinity and depth data and convert those in numeric data'''
    temp = []
    salt = []
    depth = []
    trans = []

    for lines in data:
        data_val = lines.split()
        temp.append(float(data_val[2]))
        salt.append(float(data_val[5]))
        depth.append(float(data_val[1]))
        trans.append(float(data_val[9]))
        
    return temp, salt, depth, trans
```
