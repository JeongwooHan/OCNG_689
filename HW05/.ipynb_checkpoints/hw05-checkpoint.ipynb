The boundary information for all of the states is now held in an attribute of the map object `m.states`, which is a list of lists of x-y pairs denoting the boundaries. Next use the `m.gcpoints` function to get the points along the great circle path. Convert all of these things into Shapely objects, and check for intersections. Add a buffer of 100km to the flight path to see what states you might potentially see along the way. Use this analysis to provide the following:

1. Create and display a map that shows the flight path, with the states that are seen along the flight path clearly highlighted in some way.

2. Use the information in m.states_info to extract and print a list containing the names of the states seen, with unique names. (see `np.unique`)


```python
from shapely.geometry import LineString
from shapely.geometry import Point
from shapely.geometry import Polygon
from matplotlib.collections import PatchCollection
from matplotlib.patches import PathPatch
from matplotlib.patches import Polygon as ppoly


lax = np.array([33.9425, -118.4081])
ewr = np.array([40.6925, -74.1686])

m = Basemap(llcrnrlon=-150.,llcrnrlat=20.,
        urcrnrlon=-50.,urcrnrlat=50.,
        rsphere=(6378137.00,6356752.3142),
        resolution='l',projection='merc',
        lat_0=40.,lon_0=-100.,lat_ts=20.)

m.readshapefile('/Users/SUN/HAN/OCNG_689/python4geosciences/materials/states_21basic/states', 'states',
                drawbounds = False)

lax_lat, lax_lon =m(lax[1], lax[0])
ewr_lat, ewr_lon =m(ewr[1], ewr[0])

gcpoints = m.gcpoints(lax[1], lax[0], ewr[1], ewr[0], npoints = 100)

xy = np.hstack((np.array(gcpoints[0]).reshape(len(gcpoints[0]),1), np.array(gcpoints[1]).reshape(len(gcpoints[1]),1)))
points_buff = shapely.geometry.MultiPoint(xy).buffer(100000)
points_buff

points_path = []
for x, y in zip(gcpoints[0], gcpoints[1]):
    points_path.append((x,y))

flight_path = LineString(points_path)

states_intersect = []
for info, states in zip(m.states_info, m.states):
    state_bd = shapely.geometry.Polygon(states)
    
    if points_buff.intersects(state_bd) == True:
        state_name = info['STATE_NAME']
        states_intersect.append(state_name)
    else:
        pass
        
state_name_intersect = np.unique(states_intersect)

patches   = []
for info, states in zip(m.states_info, m.states):
    if (info['STATE_NAME'] in state_name_intersect) == True:
        patches.append(ppoly(states))
    else:
        pass

fig = plt.figure(figsize=(14, 7))
ax = fig.add_subplot(111, axis_bgcolor='azure')
m.fillcontinents(lake_color = 'blue', zorder = 0)
m.drawmeridians(np.arange(-150, -50, 10), labels=[0, 0, 0, 1])
m.drawparallels(np.arange(20, 60, 5), labels=[1, 0, 0, 0])
m.readshapefile('/Users/SUN/HAN/OCNG_689/python4geosciences/materials/states_21basic/states', 'states')

# Plot LAX & EWR
m.scatter(lax_lat, lax_lon, marker = 'o', color = 'green', s = 70, label = 'LAX')
m.scatter(ewr_lat, ewr_lon, marker = 'o', color = 'green', s = 70, label = 'EWR')

# draw great circle route between NY and London
m.drawgreatcircle(lax[1], lax[0], ewr[1], ewr[0], linewidth=2, color='orange')
        
ax.add_collection(PatchCollection(patches, facecolor= 'm', linewidths=1.))
plt.draw()

print('The states can be seen during flight =', state_name_intersect)
```

### 2. Area in a Voronoi diagram

Given the series of points:

     np.array([[ 0.77354201,  0.34072038],
               [ 0.43807549,  0.93001924],
               [ 0.5624092 ,  0.79438991],
               [ 0.02536295,  0.46180244],
               [ 0.59533179,  0.07522803],
               [ 0.47867418,  0.70302822],
               [ 0.30371564,  0.83135442],
               [ 0.5856915 ,  0.16340903],
               [ 0.14700491,  0.60268997],
               [ 0.97098631,  0.51692456]])
               
create a Voronoi diagram, and calculate the area contained in the finite regions. What fraction of the total area is at least 0.001 unit away from any boundary? 0.01 away? And 0.1 away?


```python
from scipy import spatial

seed =   np.array([[ 0.77354201,  0.34072038],
           [ 0.43807549,  0.93001924],
           [ 0.5624092 ,  0.79438991],
           [ 0.02536295,  0.46180244],
           [ 0.59533179,  0.07522803],
           [ 0.47867418,  0.70302822],
           [ 0.30371564,  0.83135442],
           [ 0.5856915 ,  0.16340903],
           [ 0.14700491,  0.60268997],
           [ 0.97098631,  0.51692456]])

vor = spatial.Voronoi(seed)

patch_diagram = []

for indices in vor.regions:
    if not indices: continue     # check for empty regions
    if -1 in indices: continue   # region includes a vertex out of the diagram (the region goes to infinity)
    plt.fill(vor.vertices[indices, 0], vor.vertices[indices, 1], edgecolor='none', alpha=0.3)
    xy_vol = zip(vor.vertices[indices[:], 0], vor.vertices[indices[:], 1])
    diagram_poly = Polygon((xy_vol))
    
    patch_diagram.append(diagram_poly)
    
plt.plot(seed[:, 0], seed[:, 1], '.k')
plt.title('Voronoi Diagram before buffering')

area_all = []
for poly in patch_diagram:
    area_all.append(poly.area)
    
total_area = np.sum(area_all)
  

# 0.001 away
dist = 0.001

buffer_area001 = []
for poly in patch_diagram:
    buff_area = poly.buffer(dist)
    buffer_area001.append(buff_area.area)
    
area001_sum = np.sum(buffer_area001)


# 0.01 away
dist = 0.01

buffer_area01 = []
for poly in patch_diagram:
    buff_area = poly.buffer(dist)
    buffer_area01.append(buff_area.area)
    
area01_sum = np.sum(buffer_area01)

# 0.1 away
dist = 0.1

buffer_area1 = []
for poly in patch_diagram:
    buff_area = poly.buffer(dist)
    buffer_area1.append(buff_area.area)
    
area1_sum = np.sum(buffer_area1)

fraction001 = area001_sum/total_area
fraction01 = area01_sum/total_area
fraction1 = area1_sum/total_area

print('Total Area =', total_area)
print('Area of 0.001 away =', area001_sum)
print('Area of 0.01 away =', area01_sum)
print('Area of 0.1 away =', area1_sum)
print('fraction in the case of 0.001 away =',fraction001,'times bigger than total_area')
print('fraction in the case of 0.01 away =',fraction01,'times bigger than total_area')
print('fraction in the case of 0.1 away =',fraction1,'times bigger than total_area')
```

### 3. 2D interpolation

Data (found in `python4geosciences/data/MS03_L15.txt`) gathered from a field cruise out of the TAMU Oceanography department is randomly located in space (that is, unstructured) because of the nature of the measurement: the instrument is lowered into the water from a ship, and is brought up and down through the water column as the ship moves to get a sampling of the fields.

Read in the data file and plot the temperature in a `scatter` plot as a function of depth and latitude. Next, interpolate the temperature data to be on a uniform grid (to make it structured). Now, plot it using `contourf`, which requires structured data. Experiment with different spacing in your uniform grid, and with different numbers of contours to see how the results change, and choose values that provide a nice-looking image which closely resembles the original data.

```python
from scipy import interpolate
data = np.loadtxt('MS03_L15.txt', comments = '*')

lat = data[:, 0]
lon = data[:, 1]
depth = data[:, 2]
temp = data[:, 3]

x_grid, y_grid = np.mgrid[lat.min():lat.max(): 300j, depth.min():depth.max(): 300j]

coord = zip(lat,depth)
tri = spatial.Delaunay(list(coord))

interp = interpolate.LinearNDInterpolator(tri, temp)

z_grid = interp(x_grid, y_grid)
z_grid = np.ma.masked_where(np.isnan(z_grid), z_grid)

s_n = 3 
levs = range(int(np.floor(temp.min())),int(np.ceil(temp.max()))+s_n, s_n)

fig = plt.figure(figsize = (17,5))
ax1 = fig.add_subplot(1,2,1)
m1 = ax1.contourf(x_grid, y_grid, z_grid,levs, cmap='viridis')
ax1.set_xlim(lat.min(), lat.max())
ax1.set_ylim(depth.max(), depth.min())
ax1.set_xlabel('Latitude [Deg.]')
ax1.set_ylabel('Depth [m]')
ax1.set_title('Teperature at Uniform Grid')
fig.colorbar(m1)

ax2 = fig.add_subplot(1,2,2)
sc = ax2.scatter(lat, depth, s=100, c = temp, edgecolor = 'none', cmap='viridis')
ax2.set_xlim(lat.min(), lat.max())
ax2.set_ylim(depth.max(), depth.min())
ax2.set_xlabel('Latitude [Deg.]')
ax2.set_ylabel('Depth [m]')
ax2.set_title('Teperature Scatter Plot')
fig.colorbar(sc)
```

### 4. (Grad) Area of Greenland and Australia in different projections

Set up Basemap instances of the following projections: Mercator, Robinson, and Mollweide; make sure that all three projections are set up to show both Greenland and Australia at the same time (this is necessary for Mercator in particular). Plot them side-by-side to see how they compare.

Next, read in the shape files for Greenland and Australia, found in the `data` directory; you can do this with something like the following (but you might need to change the directory path):

> m.readshapefile('../python4geosciences/data/AUS_adm_shp/AUS_adm0', 'australia', drawbounds=False)

where `m` is a `Basemap` instance. This gives you a series of polygons that make up each country, saved in `m.greenland` and `m.australia`. Find the area of the two countries in each projection and compare them. What is the area of Australia relative to the area of Greenland for each projection, and how does this compare to what it is in reality?

```python
# Set up Basemap 
m_merc = Basemap(llcrnrlon = -150., llcrnrlat = -60.,
            urcrnrlon = 160., urcrnrlat = 85.,
            rsphere =(6378137.00,6356752.3142),
            resolution='l',projection='merc',
            lat_0=0.,lon_0=0.,lat_ts=20.)

m_robin = Basemap(llcrnrlon = -150., llcrnrlat = -60.,
            urcrnrlon = 160., urcrnrlat = 85.,
            rsphere =(6378137.00,6356752.3142),
            resolution='l',projection='robin',
            lat_0=0.,lon_0=0.,lat_ts=20.)

m_moll = Basemap(llcrnrlon = -150., llcrnrlat = -60.,
            urcrnrlon = 160., urcrnrlat = 85.,
            rsphere =(6378137.00,6356752.3142),
            resolution='l',projection='moll',
            lat_0=0.,lon_0=0.,lat_ts=20.)

fig = plt.figure(figsize=(15, 9))
ax1 = fig.add_subplot(1,3,1, axis_bgcolor='azure')
m_merc.drawcoastlines()
m_merc.fillcontinents(lake_color = 'blue', zorder = 0)

m_merc.readshapefile('GRL_adm_shp/GRL_adm0', 'greenland', drawbounds = False) 
m_merc.readshapefile('AUS_adm_shp/AUS_adm0', 'australia', drawbounds = False)

total_area_grl_merc = 0.
total_area_aus_merc = 0.

for shape_grl in m_merc.greenland:
    x_g, y_g = zip(*shape_grl)
    m_merc.plot(x_g, y_g, marker=None,color='m')
    total_area_grl_merc += shapely.geometry.Polygon(shape_grl).area
    
for shape_aus in m_merc.australia:
    x_a, y_a = zip(*shape_aus)
    m_merc.plot(x_a, y_a, marker=None,color='m')
    total_area_aus_merc += shapely.geometry.Polygon(shape_aus).area

#m.drawmeridians(np.arange(-150, 160, 20), labels=[0, 0, 0, 1])
#m.drawparallels(np.arange(-60, 85, 15), labels=[1, 0, 0, 0])
ax1.set_title('Mercator Projection')

print('Area of Green Land on mercator = ', total_area_grl_merc*10**-6, 'km^2')
print('Area of Australia on mercator= ', total_area_aus_merc*10**-6, 'km^2')

total_area_grl_robin = 0.
total_area_aus_robin = 0.

ax2 = fig.add_subplot(1,3,2, axis_bgcolor='azure')
m_robin.drawcoastlines()
m_robin.fillcontinents(lake_color = 'blue', zorder = 0)

m_robin.readshapefile('GRL_adm_shp/GRL_adm0', 'greenland', drawbounds = False) 
m_robin.readshapefile('AUS_adm_shp/AUS_adm0', 'australia', drawbounds = False)

for shape_grl in m_robin.greenland:
    x_g, y_g = zip(*shape_grl)
    m_merc.plot(x_g, y_g, marker=None,color='m')
    total_area_grl_robin += shapely.geometry.Polygon(shape_grl).area
    
for shape_aus in m_robin.australia:
    x_a, y_a = zip(*shape_aus)
    m_merc.plot(x_a, y_a, marker=None,color='m')
    total_area_aus_robin += shapely.geometry.Polygon(shape_aus).area
    
#m.drawparallels(np.arange(-90.,120.,30.), labels=[1, 0, 0, 0])
#m.drawmeridians(np.arange(0.,360.,60.), labels=[0, 0, 0, 1])
ax2.set_title('Robinson Projection')
print('Area of Green Land on robinson = ', total_area_grl_robin*10**-6, 'km^2')
print('Area of Australia on robinson= ', total_area_aus_robin*10**-6, 'km^2')

total_area_grl_moll = 0.
total_area_aus_moll = 0.

ax3 = fig.add_subplot(1,3,3, axis_bgcolor='azure')
m_moll.drawcoastlines()
m_moll.fillcontinents(lake_color = 'blue', zorder = 0)

m_moll.readshapefile('GRL_adm_shp/GRL_adm0', 'greenland', drawbounds = False) 
m_moll.readshapefile('AUS_adm_shp/AUS_adm0', 'australia', drawbounds = False)

for shape_grl in m_moll.greenland:
    x_g, y_g = zip(*shape_grl)
    m_moll.plot(x_g, y_g, marker=None,color='m')
    total_area_grl_moll += shapely.geometry.Polygon(shape_grl).area
    
for shape_aus in m_moll.australia:
    x_a, y_a = zip(*shape_aus)
    m_moll.plot(x_a, y_a, marker=None,color='m')
    total_area_aus_moll += shapely.geometry.Polygon(shape_aus).area

#m.drawparallels(np.arange(-90.,120.,30.), labels=[1, 0, 0, 0])
#m.drawmeridians(np.arange(0.,360.,60.), labels=[0, 0, 0, 1])
ax3.set_title('Mollweide Projection')
print('Area of Green Land on mollweide = ', total_area_grl_moll*10**-6, 'km^2')
print('Area of Australia on mollweide= ', total_area_aus_moll*10**-6, 'km^2')
```
